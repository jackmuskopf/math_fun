```python
import os
from PIL import Image
import numpy as np

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
image_dir = 'images'
standard_dim = (100, 100)
n_layers = 5
```

```python
# N scalars in each resized image: L x W x 3 (3 for RBG)
nd = np.multiply(*standard_dim)*3
```

```python
# read images

n_imgs = len(os.listdir(image_dir))
img_arrs = np.zeros((n_imgs,nd))
for i, fname in enumerate(os.listdir(image_dir)):
    
    image_path = os.path.join(image_dir, fname)
    
    image = Image.open(image_path)

    image = image.resize(standard_dim)

    arr = np.array(image)
    arr = arr.reshape((nd,))
    
    img_arrs[i, :] = arr

```

```python
image
```

```python
# build nn

model = keras.Sequential()
model.add(keras.Input(shape=(nd,)))
    
for _ in range(n_layers):
    model.add(layers.Dense(nd, activation='relu'))
    
model.add(layers.Dense(nd, activation='linear'))

model.compile(loss='mean_squared_error', optimizer='adam')
```
